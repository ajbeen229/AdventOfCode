# Part One

```python
import re
```

```python
def getPuzzleInput():
	with open("Day 16.txt") as file:
		return file.read()
```

```python
puzzleInput = getPuzzleInput().split('\n')

realSue = {
	"children": 3,
	"cats": 7,
	"samoyeds": 2,
	"pomeranians": 3,
	"akitas": 0,
	"vizslas": 0,
	"goldfish": 5,
	"trees": 3,
	"cars": 2,
	"perfumes": 1
}

allSues = {}

for line in puzzleInput:
	sueNumber = line.split(":")[0].split(" ")[1]
	attributes = re.findall(r'(\w+): (\d+)(?!:)', line)
	allSues[sueNumber] = {attr[0]: int(attr[1]) for attr in attributes}
	
# print(allSues)

for sue, attributes in allSues.items():
	if all([realSue[attr] == attributes[attr] for attr in attributes.keys()]):
		print(sue)
```

# Part Two

```python
reallyRealSue = {
    "children": lambda x: x == 3,
	"cats": lambda x: x > 7,
	"samoyeds": lambda x: x == 2,
	"pomeranians": lambda x: x < 3,
	"akitas": lambda x: x == 0,
	"vizslas": lambda x: x == 0,
	"goldfish": lambda x: x < 5,
	"trees": lambda x: x > 3,
	"cars": lambda x: x == 2,
	"perfumes": lambda x: x == 1   
}

# print(allSues)

for sue, attributes in allSues.items():
    if all([reallyRealSue[attr](attributes[attr]) for attr in attributes.keys()]):
        print(sue)
```
