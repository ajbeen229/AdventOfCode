# Part 1

```python
import re
from pprint import pprint
```

```python
def getPuzzleInput():
	with open("Day 14.txt") as file:
		return file.read()
```

```python
puzzleInput = getPuzzleInput().split("\n")

reindeer = {line.split(" ")[0]: list(map(int, re.findall(r'(\d+)', line))) for line in puzzleInput}

# print(reindeer)

duration = 2503

furthestDistance = (None, 0)

for name, values in reindeer.items():
	speed = values[0]
	flying = values[1]
	resting = values[2]

	# The time of one interval
	interval = flying + resting

	# How many full intervals the reindeer can complete in the duration
	totalIntervals = duration // interval
	
	# how many seconds are left after all full intervals 
	leftoverSeconds = duration % interval

	# print("Total:", totalIntervals)
	# print("Leftover:", leftoverSeconds)
	# print()

	leftoverFlying = 0
	if leftoverSeconds <= flying:
		leftoverFlying = speed * leftoverSeconds
	elif leftoverSeconds > flying:
		leftoverFlying = speed * flying
	else:
		leftoverFlying = 0

	totalDistance = speed * flying * totalIntervals + leftoverFlying

	if totalDistance > furthestDistance[1]:
		furthestDistance = (name, totalDistance)


print(f"{furthestDistance[0]} flew the farthest at {furthestDistance[1]} kilometers.")
```

# Part 2

```python
puzzleInput = getPuzzleInput().split("\n")

def getDistance(duration, speed, flying, resting):
	# The time of one interval
	interval = flying + resting

	# How many full intervals the reindeer can complete in the duration
	totalIntervals = duration // interval
	
	# how many seconds are left after all full intervals 
	leftoverSeconds = duration % interval

	# Calculate how many more seconds of flying are left after all full intervals
	leftoverFlying = 0
	if leftoverSeconds <= flying:
		leftoverFlying = speed * leftoverSeconds
	elif leftoverSeconds > flying:
		leftoverFlying = speed * flying
	else:
		leftoverFlying = 0

	return speed * flying * totalIntervals + leftoverFlying

reindeer = {
	line.split(" ")[0]: {
		"values": list(map(int, re.findall(r'(\d+)', line))),
		"distance": getDistance,
		"points": 0
	} 
	for line in puzzleInput
}

for seconds in range(1, duration+1):
	distances = []
	for name in reindeer.keys():
		distances.append((name, reindeer[name]["distance"](seconds, *reindeer[name]["values"])))

	furthestReindeer = [d for d in distances if d[1] == max([i[1] for i in distances])]

	for f in furthestReindeer:
		reindeer[f[0]]["points"] += 1


# I tried for way too long to get this working, I love my walrus
print("The reindeer with the most points is",
	  bestReindeer := max(
		nameAndPoints := {key: value["points"] for key,value in reindeer.items()},
		key=nameAndPoints.get),
	"with",
	reindeer[bestReindeer]["points"],
	"points."
)
```
