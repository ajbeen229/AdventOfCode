# Part 1

```python
from collections import defaultdict
from itertools import permutations
from pprint import pprint
import re
```

```python
def getPuzzleInput():
    with open("Day 13.txt") as file:
        return file.read()
```

```python
puzzleInput = getPuzzleInput().split("\n")

people = sorted({line.split(" ")[0] for line in puzzleInput})

seatingArrangements = list(permutations(people))

happinessCombos = defaultdict(dict)

for line in puzzleInput:
    person1 = line.split(" ")[0]
    person2 = line.split(" ")[-1][:-1]
    happiness = re.search(r'(gain|lose) (\d+)', line).groups()
    happinessUnit = -int(happiness[1]) if happiness[0] == 'lose' else int(happiness[1])

    happinessCombos[person1][person2] = happinessUnit

maximumHappiness = 0
for arrangement in seatingArrangements:
    currentHappiness = 0
    for i in range(-1, len(arrangement) - 1):
        h1 = happinessCombos[arrangement[i]][arrangement[i+1]]
        h2 = happinessCombos[arrangement[i+1]][arrangement[i]]
        currentHappiness += (h1 + h2)
    
    if currentHappiness > maximumHappiness:
        maximumHappiness = currentHappiness

print(maximumHappiness)
```

# Part 2

```python
people.append("Me")

newSeatingArrangements = list(permutations(people))

maximumHappiness = 0
for arrangement in newSeatingArrangements:
    currentHappiness = 0
    for i in range(-1, len(arrangement) - 1):
        if arrangement[i] != "Me" and arrangement[i+1] != "Me":
            h1 = happinessCombos[arrangement[i]][arrangement[i+1]]
            h2 = happinessCombos[arrangement[i+1]][arrangement[i]]
            currentHappiness += (h1 + h2)
    
    if currentHappiness > maximumHappiness:
        maximumHappiness = currentHappiness

print(maximumHappiness)
```
