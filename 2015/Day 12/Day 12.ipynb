```python
import json
import re

def getJSONPuzzleInput():
    with open("Day 12.json") as file:
        return json.load(file)

def getPuzzleInput():
    with open("Day 12.txt") as file:
        return file.read()
```

# Part 1

```python
print(sum(map(int, re.findall(r'-?\d+', getPuzzleInput()))))
```

# Part 2

```python
puzzleInput = getJSONPuzzleInput()

sum = 0

def getNextLevel(data):
    if isinstance(data, dict):
        if "red" not in data.values():
            for _, value in data.items():
                getNextLevel(value)
    elif isinstance(data, list):
        for item in data:
            getNextLevel(item)
    else:
        if isinstance(data, int):
            print(data, type(data))
            global sum
            sum += data


getNextLevel(puzzleInput)
print(sum)
```
