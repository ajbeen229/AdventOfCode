```python
from itertools import groupby
def getPuzzleInput():
    with open("Day 10.txt") as file:
        return file.read()
```

```python
def runSequence(sequence, iterations):
    newSequence = ""
    for _ in range(iterations):
        newSequence = "".join([
            x for y in list(
                map(
                    lambda x: (str(len(list(x[1]))), x[0]), groupby(sequence)
                )
            ) for x in y
        ])
        sequence = newSequence
    return len(sequence)
```

```python
sequence = getPuzzleInput()

print(runSequence(sequence, 40))
```

```python
sequence = getPuzzleInput()
print(runSequence(sequence, 50))
```
