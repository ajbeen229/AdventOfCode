```python
import string
import re
def getPuzzleInput():
    with open("Day 11.txt") as file:
        return file.read()
```

```python
oldPassword = list(getPuzzleInput())

def getNextPassword(oldPassword):
    letters = string.ascii_lowercase
    letterStraights = ["".join([letters[l], letters[l+1], letters[l+2]]) for l in range(len(letters)-2)]

    newPassword = [x for x in oldPassword]
    foundNewPassword = False

    while not foundNewPassword:
        for i in range(len(oldPassword) - 1, -1, -1):
            newPassword[i] = letters[(letters.index(oldPassword[i]) + 1) % len(letters)]
            if oldPassword[i] != 'z':
                break

        if not any(['i' in newPassword, 'o' in newPassword, 'l' in newPassword]):
            pairs = list(map(lambda x: x[0], re.findall(r'((\w)\2)', "".join(newPassword))))
            if len(set(pairs)) != 1 and len(pairs) >= 2:
                if any([straight in "".join(newPassword) for straight in letterStraights]):
                    foundNewPassword = True
                    break

        oldPassword = [x for x in newPassword]

    return ("".join(newPassword))

print(getNextPassword(oldPassword))
```
