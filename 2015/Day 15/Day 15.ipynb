```python
from itertools import product
import re
```

```python
def getPuzzleInput():
	with open("Day 15.txt") as file:
		return file.read()
```

```python
puzzleInput = getPuzzleInput().split("\n")

class Ingredient:
	allIngredients = {}
	ingredientCombos = [prod for prod in product(range(1, 100), repeat=4) if sum(prod) == 100]

	def __init__(self, name, a, cap, d, f, t, cal):
		self.name = name
		self.capacity = cap
		self.durability = d
		self.flavor = f
		self.texture = t
		self.calories = cal
		self.amount = a
		Ingredient.allIngredients[name] = self

	def __str__(self):
		return f"Cap:{self.capacity}, D:{self.durability}, F:{self.flavor}, T:{self.texture}, Cal:{self.calories}, A:{self.amount}"

	@classmethod
	def calculateCalories(cls):
		return sum([i.calories * i.amount for i in cls.allIngredients.values()])

	@classmethod
	def cookieScore(cls):
		cScore = sum([i.capacity * i.amount for i in cls.allIngredients.values()])
		dScore = sum([i.durability * i.amount for i in cls.allIngredients.values()])
		fScore = sum([i.flavor * i.amount for i in cls.allIngredients.values()])
		tScore = sum([i.texture * i.amount for i in cls.allIngredients.values()])
		if any([score <= 0 for score in [cScore, dScore, fScore, tScore]]):
			return 0
		return cScore * dScore * fScore * tScore
	
	@classmethod
	def calculateAllCookieScores(cls):
		for combo in cls.ingredientCombos:
			# Update each ingredient amount
			for i, key in enumerate(cls.allIngredients.keys()):
				cls.allIngredients[key].amount = combo[i]

			# Calculate the score of the new cookie and yield to loop
			yield cls.cookieScore()

	@classmethod
	def calculateAllCookieScoresPart2(cls):
		for combo in cls.ingredientCombos:
			# Update each ingredient amount
			for i, key in enumerate(cls.allIngredients.keys()):
				cls.allIngredients[key].amount = combo[i]

			# Calculate the score of the new cookie and yield to loop
			if cls.calculateCalories() == 500:
				yield cls.cookieScore()


# list comprehension to generate each ingredient for the cookie
[
	Ingredient(
		line.split(":")[0], # name
		25, # a
		*map(
			int, 
			[ # cap, d, f, t, cal
				prop[0], 
				prop[1], 
				prop[2], 
				prop[3], 
				prop[4]
			]
		)
	) 
	for line in puzzleInput
	if (prop := re.findall(r'-?\d+', line)) != None
]

bestScore = 0
for score in Ingredient.calculateAllCookieScores():
	# print(score)
	if score > bestScore:
		bestScore = score

print(f"The best combination of ingredients was {[f"{i.amount} {i.name}" for i in Ingredient.allIngredients.values()]} with {bestScore} points")
```

```python
bestScore = 0
for score in Ingredient.calculateAllCookieScoresPart2():
	if score > bestScore:
		bestScore = score

print(f"The best combination of ingredients with 500 calories was {[f"{i.amount} {i.name}" for i in Ingredient.allIngredients.values()]} with {bestScore} points")
```
