# Part 1

```python
import re
def getPuzzleInput():
    with open("Day 7.txt") as file:
        return file.read()
```

```python
puzzleInput = getPuzzleInput().split("\n")

inside = False
ipCount = 0

for i, line in enumerate(puzzleInput):
    supported = False
    # print(line)
    for char in range(len(line)-3):
        if line[char] == "[":
            inside = True
            # print("inside")
            continue
        elif line[char] == "]":
            inside = False
            # print("outside")
            continue

        # check for four-letter palindrome inside of brackets
        truePalin = all([
            line[char:char+2] == line[char+2:char+4][::-1],
            line[char:char+2] != line[char+2:char+4] # shouldn't be four of the same letter
        ])

        # print(char, line[char:char+4])

        if truePalin:
            if inside:
                supported = False
                # print(char, line[char:char+4], "unsupported")
                break
            else:
                supported = True
                # print(char, line[char:char+4], "supported")
    
    if supported:
        ipCount += 1

    # print()

print(ipCount)
```

```python
print("hi")
puzzleInput = getPuzzleInput().split("\n")

inside = False
ipCount = 0

for i, line in enumerate(puzzleInput):
    ABAs = []
    BABs = []
    # print(line)
    for char in range(len(line)-2):
        if line[char] == "[":
            inside = True
            # print("inside")
            continue
        elif line[char] == "]":
            inside = False
            # print("outside")
            continue
        elif "[" in line[char:char+3] or "]" in line[char:char+3]:
            continue

        # check for four-letter palindrome inside of brackets
        truePalin = all([
            line[char:char+2] == line[char+1:char+3][::-1],
            line[char:char+2] != line[char+1:char+3] # shouldn't be the same three letters
        ])

        # print(char, line[char:char+3])

        if truePalin:
            if inside:
                BABs.append(line[char:char+3])
                # print(char, line[char:char+3], "adding bab")
            elif not inside:
                ABAs.append(line[char:char+3])
                # print(char, line[char:char+3], "adding aba")
    
    print(ABAs)
    print(BABs)

    for bab in BABs:
        if f"{bab[1]}{bab[0]}{bab[1]}" in ABAs:
            ipCount += 1
            print("supported")

    print()

print(ipCount)
print("hello")
```
