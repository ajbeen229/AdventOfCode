# Part 1

```python
import numpy as np
from collections import Counter

def getPuzzleInput():
    with open("Day 8.txt") as file:
        return file.read()
```

```python
puzzleInput = getPuzzleInput().split("\n")

screen = np.full((6, 50), ".")

for line in puzzleInput:
    allIndices = []
    if line.split(" ")[0] == "rect":
        row = int(line.split(" ")[1].split("x")[1])
        col = int(line.split(" ")[1].split("x")[0])
        for r in range(row):
            for c in range(col):
                screen[r][c] = "#"
    else:
        axis = line.split(" ")[1] # row or column
        shift = list(map(int, line.split("=")[1].split(" by ")))
        if axis == "row":
            allIndices = [col for col in range(len(screen[0])) if screen[shift[0]][col] == "#"]
            newIndices = list(map(lambda x: (x + shift[1]) % len(screen[0]), allIndices))
            screen[shift[0], :] = "."
            screen[shift[0], newIndices] = "#"
        else:
            allIndices = [row for row in range(len(screen)) if screen[row][shift[0]] == "#"]
            newIndices = list(map(lambda x: (x + shift[1]) % len(screen), allIndices))
            screen[:, shift[0]] = "."
            screen[newIndices, shift[0]] = "#"

print(Counter(screen.flatten())["#"])
```
