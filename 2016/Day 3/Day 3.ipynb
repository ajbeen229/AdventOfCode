# Part 1

```python
import re

def getPuzzleInput():
    with open("Day 3.txt") as file:
        return file.read()
```

```python
puzzleInput = getPuzzleInput().split("\n")
validTriangles = 0
for sides in puzzleInput:
    side1, side2, side3 = sorted(map(int, re.findall(r'(\d+)', sides)))
    if side1 + side2 > side3:
        validTriangles += 1
print(validTriangles)
```

# Part 2

```python
newPuzzleInput = [re.findall(r'(\d+)', row) for row in getPuzzleInput().split("\n")]
newValidTriangles = 0

for col in range(3):
    for row in range(2, len(newPuzzleInput), 3):
        side1, side2, side3 = sorted(map(int, [
            newPuzzleInput[row-2][col],
            newPuzzleInput[row-1][col],
            newPuzzleInput[row][col]
        ]))

        if side1 + side2 > side3:
            newValidTriangles += 1

print(newValidTriangles)
```
