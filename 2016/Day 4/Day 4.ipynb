# Part 1

```python
import re
from collections import Counter
import string

def getPuzzleInput():
    with open("Day 4.txt") as file:
        return file.read()
```

```python
puzzleInput = getPuzzleInput().split("\n") 

realRooms = []
sectorSum = 0

for room in puzzleInput:
    name = ''.join(re.findall(r'(\w+)(?=-)', room))
    sectorID = int(re.search(r'(?<=-)(\d+)', room).group(1))
    checksum = re.search(r'\[(\w+)\]', room).group(1)
    
    letterCounts = sorted(Counter(name).items(), key=lambda x: (-x[1],x[0]))[:5]
    
    if ''.join([i[0] for i in letterCounts]) == checksum:
        sectorSum += sectorID
        realRooms.append({'name': name, 'sectorID': sectorID})

print(sectorSum)
    
```

# Part 2!

```python
def decrypt(name, sectorID):
    letters = list(string.ascii_lowercase)
    newName = ""
    for char in name:
        newLetter = letters[(letters.index(char) + sectorID) % len(letters)]
        newName += newLetter
    return newName
```

```python
for room in realRooms:
    decrypted = decrypt(room['name'], room['sectorID'])
    if re.match(r'northpoleobjects', decrypted):
        print(room['sectorID'])
        break
```
