# Part 1

```python
import numpy as np
from collections import Counter

def getPuzzleInput():
    with open("Day 6.txt") as file:
        return file.read()
```

```python
puzzleInput = [list(line) for line in getPuzzleInput().split("\n")]
charArray = np.array(puzzleInput)

decodedString = ""

for col in range(len(charArray[0])):
    # select only one column at a time to count frequency
    count = Counter(charArray[:, col])
    # print(count, "\n", list(count.keys()), "\n\n")
    decodedString += str(count.most_common(1)[0][0])

print(decodedString)
```

```python
decodedString = ""

for col in range(len(charArray[0])):
    # select only one column at a time to count frequency
    count = Counter(charArray[:, col])
    decodedString += str(count.most_common()[-1][0])

print(decodedString)
```
