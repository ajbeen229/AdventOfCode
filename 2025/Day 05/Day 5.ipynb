```python
def getPuzzleInput():
	with open("Day 5.txt") as file:
		return file.read()
```

```python
idRanges = [tuple([*map(int, line.split("-"))]) for line in getPuzzleInput().split("\n\n")[0].split("\n")]
availableIDs = map(int, getPuzzleInput().split("\n\n")[1].split("\n"))

freshIngredients = 0

for id in availableIDs:
	for line in idRanges:
		if id > line[0] and id < line[1] + 1:
			freshIngredients += 1
			break
		
print(f"Total number of fresh ingredients: {freshIngredients}")
```

```python
idRanges = [tuple([*map(int, line.split("-"))]) for line in getPuzzleInput().split("\n\n")[0].split("\n")]

allFreshIDs = 0

idRanges = sorted(idRanges, key=lambda x: x[0])

# get rid of all ranges that are enveloped in another range
i = -1
while (i := i + 1) < len(idRanges):
	if i+1 == len(idRanges):
		break
	if 	(idRanges[i+1][0] < idRanges[i][1] and idRanges[i+1][1] < idRanges[i][1]) or \
		(idRanges[i][0] == idRanges[i+1][0] and idRanges[i][1] == idRanges[i+1][1]):
		idRanges.pop(i+1)
		i -= 1

i = -1
while (i := i + 1) < len(idRanges):
	# at the last range so we can just take the difference
	if i+1 == len(idRanges):
		allFreshIDs += idRanges[i][1] - idRanges[i][0] + 1
		break

	# check for overlap between the last id of the current element and the first id of the next element
	if idRanges[i+1][0] <= idRanges[i][1]:
		overlap = idRanges[i][1] - idRanges[i+1][0]
		allFreshIDs += idRanges[i][1] - idRanges[i][0] - overlap
	else:
		allFreshIDs += idRanges[i][1] - idRanges[i][0] + 1

print(f"Total number of fresh IDs: {allFreshIDs}")
```
