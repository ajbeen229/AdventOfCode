# Part One

```python
def getPuzzleInput():
	with open("Day 4.txt") as file:
		return file.read()
```

```python
puzzleInput = [[col for col in row] for row in getPuzzleInput().split("\n")]
copy = [[y for y in x] for x in puzzleInput]

adjacentPositions = [
	lambda x, y: 1 if x-1 >= 0 and puzzleInput[x-1][y] == '@' else 0, 												# up
	lambda x, y: 1 if x-1 >= 0 and y+1 < len(puzzleInput[0]) and puzzleInput[x-1][y+1] == '@' else 0, 				# up right
	lambda x, y: 1 if y+1 < len(puzzleInput[0]) and puzzleInput[x][y+1] == '@' else 0, 								# right
	lambda x, y: 1 if x+1 < len(puzzleInput) and y+1 < len(puzzleInput[0]) and puzzleInput[x+1][y+1] == '@' else 0, # down right
	lambda x, y: 1 if x+1 < len(puzzleInput) and puzzleInput[x+1][y] == '@' else 0, 					# down
	lambda x, y: 1 if x+1 < len(puzzleInput) and y-1 >= 0 and puzzleInput[x+1][y-1] == '@' else 0, 					# down left
	lambda x, y: 1 if y-1 >= 0 and puzzleInput[x][y-1] == '@' else 0, 												# left
	lambda x, y: 1 if x-1 >= 0 and y-1 >= 0 and puzzleInput[x-1][y-1] == '@' else 0  								# up left
]

totalRolls = 0
for row in range(len(puzzleInput)):
	for col in range(len(puzzleInput[0])):
		if puzzleInput[row][col] == "@":
			if sum([x(row, col) for x in adjacentPositions]) < 4:
				copy[row][col] = 'X'
				totalRolls += 1

print(totalRolls)
```

# Part Two

```python
puzzleInput = [[col for col in row] for row in getPuzzleInput().split("\n")]
copy = [[y for y in x] for x in puzzleInput]

# check that each adjacent position is in bounds and whether or not it's another paper roll
adjacentPositions = [
	lambda x, y: 1 if x-1 >= 0 and puzzleInput[x-1][y] == '@' else 0, 												# up
	lambda x, y: 1 if x-1 >= 0 and y+1 < len(puzzleInput[0]) and puzzleInput[x-1][y+1] == '@' else 0, 				# up right
	lambda x, y: 1 if y+1 < len(puzzleInput[0]) and puzzleInput[x][y+1] == '@' else 0, 								# right
	lambda x, y: 1 if x+1 < len(puzzleInput) and y+1 < len(puzzleInput[0]) and puzzleInput[x+1][y+1] == '@' else 0, # down right
	lambda x, y: 1 if x+1 < len(puzzleInput) and puzzleInput[x+1][y] == '@' else 0, 					            # down
	lambda x, y: 1 if x+1 < len(puzzleInput) and y-1 >= 0 and puzzleInput[x+1][y-1] == '@' else 0, 					# down left
	lambda x, y: 1 if y-1 >= 0 and puzzleInput[x][y-1] == '@' else 0, 												# left
	lambda x, y: 1 if x-1 >= 0 and y-1 >= 0 and puzzleInput[x-1][y-1] == '@' else 0  								# up left
]

rollsPerRound = 0
totalRolls = 0
while True:
    rollsPerRound = 0
    # make a copy of the input each new round so we don't make changes mid round and miscount
    puzzleInput = [[y for y in x] for x in copy]
    for row in range(len(puzzleInput)):
        for col in range(len(puzzleInput[0])):
            if puzzleInput[row][col] == "@":
                # if the number of paper rolls around the current location is less than 4, count it
                if sum([x(row, col) for x in adjacentPositions]) < 4:
                    copy[row][col] = 'X'
                    rollsPerRound += 1
            
    totalRolls += rollsPerRound

    if rollsPerRound == 0:
        break

print(totalRolls)
```
