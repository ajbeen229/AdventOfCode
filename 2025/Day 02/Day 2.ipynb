```python
import re
```

```python
def getPuzzleInput():
	with open("Day 2.txt") as file:
		return file.read()
```

```python
puzzleInput = getPuzzleInput().split(",")

ranges = [str(a) for x in getPuzzleInput().split(',') for a in list(range(int(x.split('-')[0]), int(x.split('-')[1])+1))]

idSum = 0

for id in ranges:
	if len(id) % 2 == 0:
		if id[:len(id)//2] == id[len(id)//2:]:
			idSum += int(id)

print(idSum)
```

```python
idSum = 0

for id in ranges:
    if (r := re.match(r'^((\d+)\2+)$', id)) != None:
        idSum += int(id)

print(idSum)
```
