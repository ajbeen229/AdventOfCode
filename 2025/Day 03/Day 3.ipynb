# Part One 

```python
from itertools import combinations
```

```python
def getPuzzleInput():
	with open("Day 3.txt") as file:
		return file.read()
```

```python
# Expanded
puzzleInput = getPuzzleInput().split("\n")

joltage = 0

for bank in puzzleInput:
	joltageCombos = combinations(bank, 2)
	maxBankJoltage = max(joltageCombos)
	joltage += int(''.join(maxBankJoltage))

print(f"Total Joltage: {joltage}")
```

```python
# Condensed
print(f"Total Joltage: {sum([int(max(map(lambda x: ''.join([x[0], x[1]]), combinations(line, 2)))) for line in getPuzzleInput().split("\n")])}")
```

# Part Two

```python
puzzleInput = getPuzzleInput().split("\n")


def getHighestNumberFromIndex(bank, index, i):
    # if we only have one number left to get, just return the max from the rest of the list
    if i == 1:
        return max(bank[index:]), 0
    
    # otherwise, get the next best number from the rest of the list
    bestNumber = max(bank[index:(i-1)*-1])
    
    # now we need to return the next best number and the index where we found it so we only search for numbers after it
    return bestNumber, bank[index:].index(bestNumber) + index + 1
    
    
totalJoltage = 0
index = 0
for bank in puzzleInput:
    batteries = []

    # We need to find twelve numbers, so iterate over each bank twelve times
    # Iterating backwards so we dont find a number close enough to the end where
    # there aren't enough numbers after it to complete our joltage
    for i in range(12, 0, -1):
        battery, index = getHighestNumberFromIndex(bank, index, i)
        batteries.append(battery)

    totalJoltage += int(''.join(batteries))
    
print(f"Total Joltage: {totalJoltage}")
```
