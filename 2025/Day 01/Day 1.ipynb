# Part One

```python
def getPuzzleInput():
    with open("Day 1.txt") as file:
        return file.read()
```

```python
puzzleInput = getPuzzleInput().split("\n")

currentPosition = 50
count = 0

for line in puzzleInput:
    direction = line[0]
    rotation = int(line[1:]) if direction == 'R' else -int(line[1:])

    if (currentPosition + rotation) % 100 == 0:
        count += 1

    currentPosition = (currentPosition + rotation) % 100
    
print(count)
```

# Part Two

```python
currentPosition = 50
zeroClick = 0

for line in puzzleInput:
    direction = 1 if line[0] == 'R' else -1
    rotation = int(line[1:]) * direction

    iterator = iter(list(range(currentPosition + direction, currentPosition + rotation + direction, direction)))

    try:
        while True:
            if next(iterator) % 100 == 0:
                zeroClick += 1
    except:
        pass

    currentPosition += rotation

print(zeroClick)
```
