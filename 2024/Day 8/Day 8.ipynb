The signal only applies its nefarious effect at specific antinodes based on the resonant frequencies of the antennas. In particular, an antinode occurs at any point that is perfectly in line with two antennas of the same frequency - but only when one of the antennas is twice as far away as the other. This means that for any pair of antennas with the same frequency, there are two antinodes, one on either side of them.


Antennas with different frequencies don't create antinodes; A and a count as different frequencies. However, antinodes can occur at locations that contain antennas. Because the topmost A-frequency antenna overlaps with a 0-frequency antinode, there are 14 total unique locations that contain an antinode within the bounds of the map.

Calculate the impact of the signal. **How many unique locations within the bounds of the map contain an antinode**?

````python
from collections import defaultdict
from itertools import combinations
````

````python
def getPuzzleInput():
    with open('Day 8.txt') as file:
        return file.read()
````

````python
def getAntennas(puzzleInput):
    antennas = defaultdict(list)
    for row in range(len(puzzleInput)):
        for col in range(len(puzzleInput[0])):
            if puzzleInput[row][col] != '.':
                antennas[puzzleInput[row][col]].append([row, col])
    return antennas
````

````python
from collections import defaultdict
from itertools import combinations
import numpy as np

puzzleInput = getPuzzleInput().split('\n')

antennaLocations = defaultdict(list)

for row in range(len(puzzleInput)):
    for col in range(len(puzzleInput[0])):
        if puzzleInput[row][col] != '.':
            antennaLocations[puzzleInput[row][col]].append([row, col])

# for i in antennaLocations.items(): print(i)

# turn the antenna lists into combinations of antennas, two long each
for key, li in antennaLocations.items():
    # create a new list of all the possible combinations without duplicates
    if len(li) > 1:
        antennaLocations[key] = list(combinations(li, r=2)) 

# for i in antennaLocations.items(): print(i)

arr = np.full((len(puzzleInput), len(puzzleInput[0])), '.')
# for i in antennaLocations.items(): print(i)

# now plot the antinodes
antinodes = set()
for key, combos in antennaLocations.items():
    # if key == 'Y':
    for combo in combos:
        ax = combo[0][0]
        ay = combo[0][1]
        bx = combo[1][0]
        by = combo[1][1]

        arr[ax][ay] = key
        arr[bx][by] = key

        antinodeA = (ax - (bx - ax), ay - (by - ay))
        antinodeB = (bx + (bx - ax), by + (by - ay))

        if 0 <= antinodeA[0] < len(puzzleInput) and 0 <= antinodeA[1] < len(puzzleInput[0]):
            antinodes.add(antinodeA)
            arr[antinodeA[0]][antinodeA[1]] = '#'
        if 0 <= antinodeB[0] < len(puzzleInput) and 0 <= antinodeB[1] < len(puzzleInput[0]):
            antinodes.add(antinodeB)
            arr[antinodeB[0]][antinodeB[1]] = '#'

for line in arr:
    for char in line:
        print(char, end="")
    print()

print('Number of antinodes in the system:', len(antinodes))
````

Calculate the impact of the signal using this updated model. **How many unique locations within the bounds of the map contain an antinode**?

````python
def getAntennasPt2(puzzleInput):
    antennas = defaultdict(list)
    antennaCount = 0
    for row in range(len(puzzleInput)):
        for col in range(len(puzzleInput[0])):
            if puzzleInput[row][col] != '.':
                antennas[puzzleInput[row][col]].append([row, col])
                antennaCount += 1
    return antennas, antennaCount
````

````python
puzzleInput = getPuzzleInput().split('\n')

antennaLocations, antennaCount = getAntennasPt2(puzzleInput)

# turn the antenna lists into combinations of antennas, two long each
for key, li in antennaLocations.items():
    # create a new list of all the possible combinations without duplicates
    if len(li) > 1:
        antennaLocations[key] = list(combinations(li, r=2))

for i in antennaLocations.items(): print(i) if i[0] == 'A' else None

arr = np.full((len(puzzleInput), len(puzzleInput[0])), '.')

# now plot the antinodes
antinodes = set()
for key, combos in antennaLocations.items():
    # if key == 'A':
    for combo in combos:
        ax, ay = combo[0]
        bx, by = combo[1]

        xdelta = bx - ax
        ydelta = by - ay

        antinodeA = (ax - xdelta, ay - ydelta)
        antinodeB = (bx + xdelta, by + ydelta)

        arr[ax][ay] = key
        arr[bx][by] = key

        while 0 <= antinodeA[0] < len(puzzleInput) and 0 <= antinodeA[1] < len(puzzleInput[0]):
            if puzzleInput[antinodeA[0]][antinodeA[1]] == '.':
                antinodes.add(antinodeA)
            ax, ay = antinodeA
            arr[antinodeA[0]][antinodeA[1]] = '#'
            antinodeA = (ax - xdelta, ay - ydelta)


        while 0 <= antinodeB[0] < len(puzzleInput) and 0 <= antinodeB[1] < len(puzzleInput[0]):
            if puzzleInput[antinodeB[0]][antinodeB[1]] == '.':
                antinodes.add(antinodeB)
            bx, by = antinodeB
            arr[antinodeB[0]][antinodeB[1]] = '#'
            antinodeB = (bx + xdelta, by + ydelta)

for line in arr:
    for char in line:
        print(char, end="")
    print()


print(sorted(antinodes))
print(len(antinodes) + antennaCount)
````
